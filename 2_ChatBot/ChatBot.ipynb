# ChatBot Model

```python
# Importing Required Libraries
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
# Selecting the model for the ChatBot development
model_name = "facebook/blenderbot-400M-distill"
```

```python
# Initializing the selected model and tokenizer
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# To manintain the conversation History with the Chatbot
conversation_history = []

# Joining the COnversation History as one text to feed back to the model with the new input string
history_string = "\n".join(conversation_history)
```

## Check the generated model

```python
# Input Text to check the Chatbot response
input_text = "Hello! How are you doing?"
```

```python
# Combining the conversation history with the input text and the tokenizing it
inputs = tokenizer(history_string + "\n" + input_text, return_tensors="pt")
print(inputs)
```

```python
# Generating the output
outputs = model.generate(**inputs)
print(outputs)
```

```python
# decoding the tokenized output
response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
print(response)
```

```python
# Updating the Conversation History
conversation_history.append(input_text)
conversation_history.append(response)
print(conversation_history)
```

## Chatbot Conversation Loop

```python
while True:
  # Create conversation history string
  history_string = "\n".join(conversation_history)

  # Get the input data from the user
  input_text = input("> ")

  # Exit the loop
  if input_text.lower() in ["exit", "quit", "bye", "stop"]:
    print("Chat ended!!!")
    break

  # Tokenize the input text and history
  inputs = tokenizer(history_string + "\n" + input_text, return_tensors="pt")

  # Generate the response from the model
  outputs = model.generate(**inputs)

  # Decode the response
  response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

  # Print the response
  print(response)

  # Add the interaction to conversation history
  conversation_history.append(input_text)
  conversation_history.append(response)
```
